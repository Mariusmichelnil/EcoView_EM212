```python
import json
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

# Counties

```python
df_counties = pd.read_csv('counties.csv')
df_counties
```

# Drought

```python
# For every FIPS we import the data and concatenate the datasets with some modifications on the index and date
FIPS = df_counties.FIPS.to_list()
dfs_drought = []
for id in FIPS:
    dfs_drought.append(pd.read_json('data/'+str(id)+'_drought.json'))
all_df_drought = pd.concat(dfs_drought)
all_df_drought = all_df_drought.reset_index()
df_drought = all_df_drought.drop(['MapDate', 'ValidEnd', 'StatisticFormatID'], axis=1)
df_drought = df_drought.rename(columns={"ValidStart": "date"}, errors="raise")
df_drought['date'] = pd.to_datetime(df_drought['date'], format='%Y-%m-%d')
df_drought
```

```python
# We save the dataset
df_drought.to_csv('drought.csv')
```

# Weather

```python
# For every FIPS we import the data and concatenate the datasets with some modifications on the index and date
FIPS = df_counties.FIPS.to_list()
dfs_weather = []
for id in FIPS:
    df1 = pd.read_json('data/'+str(id)+'_weather.json')
    df1["FIPS"] = id
    df1["date"] = df1.index
    df1 = df1.reset_index()
    dfs_weather.append(df1)
all_df_weather = pd.concat(dfs_weather)
all_df_weather = all_df_weather.reset_index()
all_df_weather['date'] = pd.to_datetime(all_df_weather['date'], format='%Y%m%d')
all_df_weather.head()
```

```python
# Try to understand where the difference of number of rows comes from
df = all_df_weather[all_df_weather.FIPS == 1001]
df['gap'] = df['date'].sort_values().diff() > pd.to_timedelta('1 day')
df[df.gap]
```

Drought data is only weekly

Before merge we need to average the weather data per week.

First lets make sure we have the same date range in both datasets

```python
print(all_df_weather.date.min())
print(df_drought.date.min())
```

```python
df_drought = df_drought[df_drought['date'] >= pd.to_datetime('1/1/2018')]
df_drought.date.min()
```

```python
all_df_weather = all_df_weather[all_df_weather['date'] >= pd.to_datetime('1/2/2018')]
all_df_weather.date.min()
```

```python
df_weather = all_df_weather.copy()

#calculate sum of values, grouped by week
df_weather = df_weather.groupby(['FIPS',pd.Grouper(key='date', freq="W-TUE")]).mean()
df_weather.reset_index(inplace=True)
df_weather
```

```python
# Save the dataset
df_weather.to_csv('weather.csv')
```

# Merge

```python
# To merge we create a column of date + fips because we want to join on date and county
df_drought['date_fips'] = df_drought['date'].astype('str') + df_drought['FIPS'].astype('str')
all_df_weather['date_fips'] = all_df_weather['date'].astype('str') + all_df_weather['FIPS'].astype('str')
df_drought['date_fips'] = df_drought['date_fips'].astype('str')
all_df_weather['date_fips'] = all_df_weather['date_fips'].astype('str')
df_drought = df_drought.drop(['index'], axis=1)
all_df_weather = all_df_weather.drop(['level_0','index'], axis=1)

# We create a drought score that we will try to predict
df_drought['score'] =\
(df_drought['D0']/100)+\
(df_drought['D1']/100)+\
(df_drought['D2']/100)+\
(df_drought['D3']/100)+\
(df_drought['D4']/100)
```

```python
# check the result
all_df_weather.head()
```

```python
df_drought.head()
```

```python
# We join the datasets, weather and droughts
join_df1 = df_drought.merge(all_df_weather, on='date_fips', how= 'left')
join_df1 = join_df1.drop(['FIPS_y', 'date_y'], axis=1)
join_df1 = join_df1.rename(columns={"date_x": "date", "FIPS_x": "FIPS"}, errors="raise")
join_df1.head()
```

```python
# We add the county dataset
join_df = join_df1.merge(df_counties, on='FIPS', how= 'left')
join_df = join_df.drop(['State_y'], axis=1)
join_df = join_df.rename(columns={"State_x": "State"}, errors="raise")
join_df
```

```python
# Save the raw dataset to csv
join_df.to_csv('data_raw.csv',index=False)
```
